```python
from pathlib import Path
import sys

# Add the 'runners' directory to sys.path
sys.path.insert(0, 'runners')

from paddle_runner import PaddleRunner
```

```python
DATASET = Path("datasets/English")
OUTPUT = Path("outputs/paddleocr")

OUTPUT.mkdir(parents=True, exist_ok=True)
```

```python
runner = PaddleRunner()
```

```python
pdfs = sorted(DATASET.glob("*.pdf"))

for pdf in pdfs:

    print(f"Processing {pdf.name}")

    markdown = runner.run(pdf)

    output_file = OUTPUT / f"{pdf.stem}.md"

    with open(output_file, "w", encoding="utf-8") as f:
        f.write(markdown)
    print(f"Generated markdown for {pdf.name}: {len(markdown)} characters")
```

```python
import shutil
from google.colab import files

# Create a zip archive of the outputs folder
output_zip_path = shutil.make_archive(OUTPUT, 'zip', OUTPUT)

print(f"'{output_zip_path}' created. Downloading now...")

# Download the zip file
files.download(output_zip_path)
```
